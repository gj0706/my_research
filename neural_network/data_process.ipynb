```python
import pandas as pd
```

```python
w1 = pd.read_json('data/w1.json')
w2 = pd.read_json('data/w2.json')
w1.head()
```

```python
feature = ['feature1', 'feature2', 'feature3', 'feature4', 'feature5', 'feature6', 'feature7', 'feature8']
```

```python
w1['feature'] = feature
w1 = w1.set_index('feature').reset_index()
w1.head(8)
```

```python
def flatten_col(df, col, reset_index=False):
    col_flat = pd.DataFrame([[i, x] for i, y in df[col].apply(list).iteritems() for x in y], columns=['I', col])
    col_flat = col_flat.set_index('I')
#     df = df.drop(col, 1)
#     df = df.merge(col_flat, left_index=True, right_index=True)
#     if reset_index:
#         df = df.reset_index(drop=True)
    return col_flat
```

```python
i1_flatten = flatten_col(w1, 'i')
f1_flatten = flatten_col(w1, 'f')
c1_flatten = flatten_col(w1, 'c')
o1_flatten = flatten_col(w1, 'o')
i1_flatten
```

```python
w1_flatten = pd.concat([i1_flatten,f1_flatten,c1_flatten,o1_flatten], axis=1)
w1_flatten.reset_index(inplace=True)
w1_flatten.rename(columns={'I':'feature'},inplace=True)
w1_flatten
```

```python
cell = list(range(50)) * 8
w1_flatten['cell'] = cell
w1_flatten
```

```python
w1_melt = pd.melt(w1_flatten,id_vars=['feature','cell'], value_vars=['i','f','c','o'], var_name='gate')
w1_melt
```

```python
i2_flatten = flatten_col(w2, 'i')
f2_flatten = flatten_col(w2, 'f')
c2_flatten = flatten_col(w2, 'c')
o2_flatten = flatten_col(w2, 'o')
i2_flatten
```

```python
w2_flatten = pd.concat([i2_flatten,f2_flatten,c2_flatten,o2_flatten], axis=1)
w2_flatten.reset_index(inplace=True)
w2_flatten.rename(columns={'I':'feature'},inplace=True)
w2_flatten
```

```python
w2_flatten['feature'].dtype
```

```python
cell = list(range(50)) * 50
w2_flatten['cell'] = cell
w2_flatten
```

```python
w2_melt = pd.melt(w2_flatten,id_vars=['feature','cell'], value_vars=['i','f','c','o'], var_name='gate')
w2_melt
```

```python
w1_flatten.to_json('data/w1_flatten.json',orient='records')
```
