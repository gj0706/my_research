```python
def read_weight_file(file_path):
    f = h5py.File(file_path,'r')
    return f
```

```python
def add_feature_column(df):
    df.reset_index(inplace=True)
    df.rename(columns={'index':'feature'}, inplace=True)
    return df
```

```python
def extract_weights(file_path, n_layer ):
    f = read_weight_file(file_path)
    layer = f['model_weights/LSTM{}/LSTM{}_2/kernel:0'.format(n_layer, n_layer)]
    df = pd.DataFrame(np.array(layer))
#     df = add_feature_column(df)
    return df
```

```python
def melt_columns(df):
    df_melt = pd.melt(df, id_vars='feature', var_name='cell')
    return df_melt
```

```python
def concat_columns(df):
    add_feature_column(df)
    n_cells = len(df.columns) // 4 -1
    i = df.iloc[:, 0:(n_cells + 1)]
    f = df.iloc[:, (n_cells + 1):(n_cells * 2 + 1)]
    c = df.iloc[:, (n_cells * 2 + 1):(n_cells * 3 + 1)]
    o = df.iloc[:, (n_cells * 3 + 1):(n_cells * 4 + 1)]
    
#     add_feature_column(i)
#     add_feature_column(f)
#     add_feature_column(c)
#     add_feature_column(o)


    i_melt = melt_columns(i)
    f_melt = melt_columns(f)
    c_melt = melt_columns(c)
    o_melt = melt_columns(o)
    
    i_melt['gate']=['i' for _ in range(400)]
    f_melt['gate']=['f' for _ in range(400)]
    c_melt['gate']=['c' for _ in range(400)]
    o_melt['gate']=['o' for _ in range(400)]


   
    all_melt = pd.concat([i_melt,f_melt,c_melt,o_melt], axis=0)


    return all_melt
```
