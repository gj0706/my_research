```python
def extract_weights(file_path, n_layer ):
    f = read_weight_file(file_path)
    layer = f['model_weights/LSTM{}/LSTM{}_4/kernel:0'.format(n_layer, n_layer)]
    df = pd.DataFrame(np.array(layer))
#     df = add_feature_column(df)
    return df

n_cells = 50
def add_columns(df,gate):
    df.reset_index(inplace=True)
    df.rename(columns={'index':'cell_l1'}, inplace=True)
    df['gate'] = [gate for _ in range(50)]
    return df

def melt_columns(df):
    df_melt = pd.melt(df, id_vars=['cell_l1','gate'], var_name='cell_l2')
    return df_melt

def transform_gate_df(df,nth_epoch):
    i = df.iloc[:, 0 : n_cells]
    f = df.iloc[:, n_cells :n_cells * 2]
    c = df.iloc[:, n_cells * 2 : n_cells * 3]
    o = df.iloc[:, n_cells * 3 : n_cells * 4]
    
#     return i, f, c, o
    add_columns(i, 'i')
    add_columns(f, 'f')
    add_columns(c, 'c')
    add_columns(o, 'o')

    
    i_melt = melt_columns(i)
    f_melt = melt_columns(f)
    f_melt['cell_l2'] = i_melt['cell_l2']
    c_melt = melt_columns(c)
    c_melt['cell_l2'] = i_melt['cell_l2']
    o_melt = melt_columns(o)
    o_melt['cell_l2'] = i_melt['cell_l2']

    
    all_melt = pd.concat([i_melt,f_melt,c_melt,o_melt], axis=0)
    epoch = [nth_epoch for _ in range(10000)]
    all_melt['epoch'] = epoch
    return all_melt
#     return i_melt, f_melt, c_melt, o_melt

def concat_columns(df_list):
   
    all_melt = pd.concat(df_list, axis=0)


    return all_melt
```
